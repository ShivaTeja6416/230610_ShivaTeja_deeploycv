```python
import cv2
import matplotlib.pyplot as plt

# Load an image using OpenCV
img1 = cv2.imread('prob1_img1.jpg', 0)  # 0 for grayscale mode
img2 = cv2.imread('prob1_img2.jpg', 0)
img3 = cv2.imread('prob1_img3.jpg', 0)

# Function to display an image
def show(name, n, m, i, Title):
    plt.subplot(n, m, i)          # Divide plot into n rows and m columns
    plt.imshow(name, cmap='gray') # Display image in grayscale
    plt.title(Title)              # Add title to the image
    plt.axis('off')               # Remove the axes for cleaner display

# Display all images side by side
plt.figure(figsize=(20, 10))  # Set the size of the display window

show(img1, 1, 3, 1, "Opera house")  # First image
show(img2, 1, 3, 2, "Dog")    # Second image
show(img3, 1, 3, 3, "Human")   # Third image

plt.show()  # Show all images



```
